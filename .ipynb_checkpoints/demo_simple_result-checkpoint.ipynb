```python
%matplotlib notebook

import os
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
files = os.listdir("result")
```

```python
fig, axes = plt.subplots(nrows = 5, ncols= 4, figsize=(7, 7))
good_files = [ f for f in files if f[0] == 'g']
print(good_files)

for i in range(4):
    for j in range(5):
        ax = axes[j][i]
        path_to_file = os.path.join('result',good_files[i*5+j])
        img = Image.open(path_to_file)
        ax.imshow(img)
        ax.axis('off')
        ax.set_aspect('equal')
plt.subplots_adjust(wspace=None, hspace=None)
plt.tight_layout()

fig.canvas.draw()
        
```

```python
fig, axes = plt.subplots(nrows = 1, ncols= 3, figsize=(7, 7))
bad_files = [ f for f in files if f[0] == 'b']
print (bad_files)
for i in range(3):
    ax = axes[i]
    path_to_file = os.path.join('result',bad_files[i])
    img = Image.open(path_to_file)
    img_np = np.array(img)/255.0
    ax.imshow(img_np, )
    ax.axis('off')
# plt.tight_layout()
fig.canvas.draw()
```
