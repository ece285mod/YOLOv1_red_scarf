```python
%matplotlib notebook

import os
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
files = os.listdir("result")
```

```python
fig, axes = plt.subplots(nrows = 5, ncols= 4, figsize=(7, 7),  gridspec_kw = {'wspace':0, 'hspace':0})
good_files = [ f for f in files if f[0] == 'g']
print(good_files)

for i in range(4):
    for j in range(5):
        ax = axes[j][i]
        path_to_file = os.path.join('result',good_files[i*5+j])
        img = Image.open(path_to_file)
        ax.imshow(img)
        ax.axis('off')
        ax.set_aspect('equal')

fig.subplots_adjust(wspace=0, hspace=0)
plt.tight_layout()

fig.canvas.draw()
        
```
